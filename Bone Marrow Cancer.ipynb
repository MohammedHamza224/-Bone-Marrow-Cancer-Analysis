### 1- Country  Analysis

#### **KPI Cards**

```python
# Total Myeloma Cases
total_myeloma = country['Myeloma_New_Cases'].sum()

# Total Leukemia Cases
total_leukemia = country['Leukemia_New_Cases'].sum()

# Total Deaths
total_deaths = (
    country['Myeloma_Deaths'].sum()
    + country['Leukemia_Deaths'].sum()
)

# Average Survival
avg_survival = country['Myeloma_5Y_Survival_Pct'].mean()

# Total BMT Centers
total_bmt = country['BMT_Centers'].sum()

# Countries Count
countries = country['Country'].nunique()


print("Total Myeloma Cases:", total_myeloma)
print("Total Leukemia Cases:", total_leukemia)
print("Total Deaths:", total_deaths)
print("Average Survival %:", round(avg_survival,2))
print("BMT Centers:", total_bmt)
print("Countries:", countries)
```

```python
# if incase to filter by 5 countries to visulization the data for those countries only .......
selected_countries = country[
    country['Country'].isin([
        'Egypt',
        'Saudi Arabia',
        'Sudan',
        'United States',
        'Germany'
    ])
]

selected_countries
```

```python
# top 10 countries with the highest number of new cases of bone marrow cancer.
# if in case non Filter the data to show only the top 10 countries with the highest number of new cases ...
import matplotlib.pyplot as plt

top_myeloma = country.nlargest(10, 'Myeloma_New_Cases')

plt.figure(figsize=(10,5))
plt.bar(top_myeloma['Country'], top_myeloma['Myeloma_New_Cases'])
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Myeloma Cases')
plt.ylabel('New Cases')
plt.tight_layout()
plt.show()
```

```python
# top 10 countries with the highest number of new cases of Leukemia cancer .
top_leukemia = country.nlargest(10, 'Leukemia_New_Cases')

plt.figure(figsize=(10,5))
plt.bar(top_leukemia['Country'], top_leukemia['Leukemia_New_Cases'])
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Leukemia Cases')
plt.ylabel('New Cases')
plt.tight_layout()
plt.show()
```

```python
# top 10 countries with the highest survival rates for bone marrow cancer  
top_survival = country.nlargest(
    10,
    'Myeloma_5Y_Survival_Pct'
)

plt.figure(figsize=(10,5))
plt.bar(
    top_survival['Country'],
    top_survival['Myeloma_5Y_Survival_Pct']
)

plt.xticks(rotation=45)
plt.title('Top Survival Rate Countries')
plt.ylabel('Survival %')
plt.tight_layout()
plt.show()
```

```python
# top 10 countries with the highest number of BMT centers for bone marrow cancer 
top_bmt = country.nlargest(
    10,
    'BMT_Centers'
)

plt.figure(figsize=(10,5))
plt.bar(
    top_bmt['Country'],
    top_bmt['BMT_Centers']
)

plt.xticks(rotation=45)
plt.title('Top Countries by BMT Centers')
plt.ylabel('Centers')
plt.tight_layout()
plt.show()
```

```python
# top 10 countries with the highest number of total cases and deaths for bone marrow cancer 
country['Total_Cases'] = (
    country['Myeloma_New_Cases']
    + country['Leukemia_New_Cases']
)

country['Total_Deaths'] = (
    country['Myeloma_Deaths']
    + country['Leukemia_Deaths']
)

top = country.nlargest(
    10,
    'Total_Cases'
)

top.plot(
    x='Country',
    y=['Total_Cases','Total_Deaths'],
    kind='bar',
    figsize=(10,5)
)

plt.title('Cases vs Deaths')
plt.show()
```

```python
# Scatter plot to visualize the relationship between the number of BMT centers and the 5- years survival rate for bone marrow cancer in different countries.
plt.figure(figsize=(8,5))

plt.scatter(
    country['BMT_Centers'],
    country['Myeloma_5Y_Survival_Pct']
)

plt.xlabel('BMT Centers')
plt.ylabel('5-Year Survival %')
plt.title('BMT Centers vs Survival Rate')

plt.show()
```

```python
# Group the data by continent and sum the number of new cases for myeloma and Leukimia to visulization the distribution of bone marrow cancer.
continent_cases = country.groupby(
    'Continent'
)[['Myeloma_New_Cases',
   'Leukemia_New_Cases']].sum()

continent_cases.plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Cancer Cases by Continent')
plt.ylabel('Cases')
plt.show()
```

```python
# Heatmap to visualize the correlation between diffrence numerical variables in the country dataset.
import seaborn as sns

plt.figure(figsize=(12,8))

sns.heatmap(
    country.select_dtypes(include='number').corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Matrix')
plt.show()
```

### 3- Survival Table

```python
survival.columns
```

###  **KPI Cards**

```python
# Average survival rates

avg_1 = survival['One_Year_Survival_Pct'].mean()
avg_3 = survival['Three_Year_Survival_Pct'].mean()
avg_5 = survival['Five_Year_Survival_Pct'].mean()
avg_10 = survival['Ten_Year_Survival_Pct'].mean()

print("1 Year:", round(avg_1,2), "%")
print("3 Years:", round(avg_3,2), "%")
print("5 Years:", round(avg_5,2), "%")
print("10 Years:", round(avg_10,2), "%")
```

```python
# 5-Year Survival Rate by Cancer Type
import matplotlib.pyplot as plt

top_survival = survival.sort_values(
    'Five_Year_Survival_Pct',
    ascending=False
)

plt.figure(figsize=(10,5))

plt.bar(
    top_survival['Cancer_Type'],
    top_survival['Five_Year_Survival_Pct']
)

plt.xticks(rotation=45)
plt.title('5-Year Survival Rate by Cancer Type')
plt.ylabel('Survival %')

plt.tight_layout()
plt.show()
```

```python
# 1 Year vs 5 Year Survival Rate by Cancer Type
compare = survival[['Cancer_Type',
                    'One_Year_Survival_Pct',
                    'Five_Year_Survival_Pct']]


compare.plot(
    x='Cancer_Type',
    kind='bar',
    figsize=(10,5)
)

plt.title('1 Year vs 5 Year Survival')
plt.ylabel('Survival %')
plt.xticks(rotation=45)

plt.show()
```

```python
# 5-Year Survival Rate by Treatment Access Level    البقاء على قيد الحياة لمدة 5 سنوات حسب مستوى الوصول إلى العلاج
access = survival.groupby(
    'Treatment_Access_Level'
)['Five_Year_Survival_Pct'].mean()


access.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Survival by Treatment Access Level')
plt.ylabel('5 Year Survival %')

plt.show()
```

```python
# 5-Year Survival Rate by Income Region    البقاء على قيد الحياة لمدة 5 سنوات حسب مستوى الدخل
region = survival.groupby(
    'Income_Region'
)['Five_Year_Survival_Pct'].mean()


region.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Survival by Income Region')
plt.ylabel('5 Year Survival %')

plt.show()
```

```python
# 5-Year Survival Rate by Cancer Type, Survival Rate, and Typical Treatment  افضل معدل البقاء على قيد الحياة لمدة 5 سنوات حسب نوع السرطان ونسبة البقاء على قيد الحياة والعلاج النموذجي
survival[['Cancer_Type',
          'Five_Year_Survival_Pct',
          'Typical_Treatment']].sort_values(
              'Five_Year_Survival_Pct',
              ascending=False
          )
```
